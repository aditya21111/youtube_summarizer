### Using agents

```python
from langchain_community.document_loaders import YoutubeLoader
from langchain_community.document_loaders import UnstructuredURLLoader


def get_transcript(url:str)->str:
    """Given a youtube video url it will generate transcript required to generate summary"""
    try:
        loader=YoutubeLoader.from_youtube_url(url,add_video_info=False)
        docs=loader.load()

        return docs[0].page_content
    except Exception as e:
        print(e)
```

```python
def get_doc_from_url(url:str)->str:
    """given a non youtube url it will generate the doc needed to summarize"""
    try:
        loader=UnstructuredURLLoader(urls=[url])
        docs=loader.load()        
        return "\n".join([doc.page_content for doc in docs])
    
    except Exception as e:
        print(e)
```

```python
from langchain.agents import create_agent
from langchain_groq import ChatGroq
from langchain_google_genai import ChatGoogleGenerativeAI
#groqllm=ChatGroq(model='qwen/qwen3-32b',reasoning_format='hidden')
googlellm=ChatGoogleGenerativeAI(model='gemini-3.1-flash-lite')



agent=create_agent(
model=googlellm,
tools=[get_transcript,get_doc_from_url],
system_prompt='''you are a multilingual text expert.your job is to summarize the given content while preserving most of information.User markdown with sections to give output. ''',
)
```

```python
agent.invoke({'messages':'https://www.youtube.com/watch?v=x2VHFgyawPE'})
```
